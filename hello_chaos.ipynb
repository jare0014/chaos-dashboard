```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Physics Setup: Damped Harmonic Oscillator
t = np.linspace(0, 10, 500)  # Time from 0 to 10s
decay = np.exp(-t / 3)       # Decay envelope (e^-t/tau)
oscillation = np.sin(2 * np.pi * t) # Sine wave
physics_wave = decay * oscillation

# 2. The Plot
plt.figure(figsize=(10, 4))
plt.plot(t, physics_wave, label='Damped Oscillation', color='cyan')
plt.plot(t, decay, '--', color='red', alpha=0.5, label='Envelope')
plt.plot(t, -decay, '--', color='red', alpha=0.5)
plt.title("System Check: The Physics Engine is Online")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
